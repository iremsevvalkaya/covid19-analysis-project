```python
#Tematik Trend Analizi
keywords = {
    "Vaccine": "vaccine|vaccination|mrna",
    "Variant": "variant|mutation|strain",
    "Symptom": "symptom|clinical sign|manifestation"
}

trend_data = {}

for label, pattern in keywords.items():
    mask = (
        df_clean["title"].str.contains(pattern, case=False) |
        df_clean["abstract"].str.contains(pattern, case=False)
    )
    yearly_counts = df_clean[mask]["year"].value_counts().sort_index()
    trend_data[label] = yearly_counts

# DataFrame'e dönüştür
trend_df = pd.DataFrame(trend_data).fillna(0).astype(int)
```

```python
#Karşılaştırmalı Grafik
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
for column in trend_df.columns:
    plt.plot(trend_df.index, trend_df[column], marker='o', label=column)

plt.title("Covid-19 Yayınlarında Tematik Trendler")
plt.xlabel("Yıl")
plt.ylabel("Makale Sayısı")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

 ## AŞAMA: WordCloud – Kelime Bulutu ile Yaygın Kavramlar
Amaç:
Özet (abstract) alanında geçen en yaygın kelimeleri görsel olarak göstermek
Bilim insanlarının metinlerde hangi kavramlara daha çok odaklandığını tespit etmek
Yayınlara genel bakış sağlamak

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
import re

nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
# 1. Başlıkları birleştirdim
titles = " ".join(df_clean["title"].dropna())
```

```python
# 2. Temizleme işlemlerini yaptım
titles = re.sub(r'\d+', '', titles)  # Sayıları kaldırdım
titles = re.sub(r'[^\w\s]', '', titles)  # Noktalamayı kaldırdım
titles = titles.lower()  # Küçük harf
```

```python
# 3. Stopword'leri kaldırdım
filtered_words = [word for word in titles.split() if word not in stop_words]
filtered_text = " ".join(filtered_words)
```

```python
# 4. WordCloud görselleştirdim
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(filtered_text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Covid-19 Başlıklarında En Sık Geçen Kelimeler", fontsize=16)
plt.tight_layout()
plt.show()
```

 ## Başlık Temelli WordCloud Analizi

Bu görsel, Covid-19 literatüründeki makale başlıklarında en sık geçen terimleri ortaya koymaktadır. Başlıklarda "covid", "virus", "patients", "infection" gibi anahtar kavramlar ön plana çıkmaktadır. Bu analiz, literatürün genel yönelimi hakkında hızlı bir içgörü sunar.


 ## TF-IDF ile Anahtar Kelime Çıkarımı

Amaç:
Her bir makale özeti (abstract) için metindeki kelimelerin önemini belirlemek
Yayınlarda en bilgilendirici terimleri çıkarmak
u sayede “covid”, “virus”, “patient” gibi yaygın ama düşük ayırt ediciliği olan kelimeleri eleyip, anlamlı temaları öne çıkarmak

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
#abstract verisini temizle ve eksik olanları filtreledim
df_abstracts = df_clean.dropna(subset=["abstract"]).copy()
```

```python
#TF-IDF analizini yaptım
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF nesnesi
vectorizer = TfidfVectorizer(stop_words='english', max_features=100)

# TF-IDF matrisi
X = vectorizer.fit_transform(df_abstracts["abstract"])

# Kelimeleri ve skorlarını aldım
tfidf_scores = zip(vectorizer.get_feature_names_out(), X.sum(axis=0).tolist()[0])

# Skora göre sıraladım
sorted_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)
```

```python
import pandas as pd

# İlk 20 kelime ve skorlarını DataFrame'e aldım
top_words_df = pd.DataFrame(sorted_scores[:20], columns=["Kelime", "TF-IDF Skoru"])
print(top_words_df)
```

- TF-IDF analizi, “abstract” alanındaki en bilgilendirici kelimeleri ortaya koyar.
- “infection”, “patients”, “symptoms”, “vaccine” gibi sözcüklerin ön planda olması, yayınların klinik bulgulara ve pandemi yönetimine odaklandığını gösterir.


### TF-IDF Skorlarıyla WordCloud Görselleştirme

```python
# TF-IDF skorlarını sözlük formatına dönüştürtüm
word_scores = dict(sorted_scores[:100])  # İlk 100 kelime
```

```python
# WordCloud oluşturdum
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(width=1000, height=500, background_color="white")
wc.generate_from_frequencies(word_scores)
```

```python
# Görselleştirdim
plt.figure(figsize=(15, 7))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("TF-IDF Skorlarına Göre En Önemli 100 Kelime", fontsize=16)
plt.show()
```

- WordCloud, “abstract” alanında en çok ayırt edici kelimeleri temsil eder.
- TF-IDF skoruna göre oluşturulan bu görselde "virus", "infection", "patients", "covid" gibi kelimelerin öne çıktığı görülmektedir.
- Bu kelimeler, pandeminin klinik, tedavi ve halk sağlığı yönlerinin yayınlara nasıl yansıdığını göstermektedir.
